# Multiprocessing

### Intro, PID, etc.

This example demonstrates the difference between getting the current process ID and its parent process ID. When a Python script runs, it's itself a process, and it's usually started by another process (like your shell, a Colab runtime, or an IDE).

```python
import os
import multiprocessing
import time

# —- child process territory —-
def child_process_function():
  print(f"  [Child Process] My PID: {os.getpid()}")
  print(f"  [Child Process] My Parent PID (PPID): {os.getppid()}")

# —- master process territory —-
if __name__ == "__main__":
  print(f"[Main Script] My PID: {os.getpid()}")
  print(f"[Main Script] My Parent PID (PPID): {os.getppid()}\n")

  print("Starting a child process...")
  child_proc = multiprocessing.Process(target=child_process_function)
  child_proc.start()
  child_proc.join()
  print("\nChild process finished.")
```

### Async1 Pool

```python
import multiprocessing as mp
import time

def cube(x):
  time.sleep(0.1) # Simulate some work
  print(f"Cubing {x}: {x**3}")
  return x**3

def square(x):
  time.sleep(0.1) # Simulate some work
  print(f"Squaring {x}: {x**2}")
  return x**2

if __name__ == "__main__":
  numbers = [2, 3, 4]
  results = []

def collect_result(value):
  results.append(value)

print("Starting multiprocessing pool...")
with mp.Pool(processes=mp.cpu_count()) as pool:
# Heterogeneous tasks: cube and square
  for num in numbers:
    pool.apply_async(cube, (num,), \
    callback=collect_result)

    for num in numbers:
      pool.apply_async(square, (num,), callback=collect_result)

  print("All tasks dispatched. Waiting for results...")
  print("Master task 1: 1+1 = ", 1+1)
  pool.close() # No more tasks will be added
  print("Master task 2: 2+2 = ", 2+2)
  pool.join()  # Wait for all tasks to complete
  print("All tasks completed.")
  print(f"Collected results: {results}")
```

```python
import multiprocessing
import time

def slow_square(n):
    time.sleep(2) # heavy HPC calculation
    return n * n

if __name__ == "__main__":
    pool = multiprocessing.Pool(processes=4)

    # 1. Fire off the task.
    # 'result_obj' is the AsyncResult
    result_obj = pool.apply_async(slow_square, (5,))

    print("Main process is free to do other work.")

    # 2. Check if it's done without stopping
    print(f"Is the work finished? {result_obj.ready()}")

    # 3. Now we ask for the data.
    # WAIT until the 2 seconds are up.
    final_answer = result_obj.get()

    print(f"Work finished! Result: {final_answer}")

    pool.close()
    pool.join()
```

### Async using Callback

```python
import multiprocessing
import time

# Global list to store results as they come in
results_list = []

def slow_square(n):
    time.sleep(2) # heavy HPC calculation
    return n * n

def collect_result(value):
    """
    This is the callback function.
    It runs in the main process once a worker is done.
    """
    results_list.append(value)
    print(f"Callback received: {value}")

if __name__ == "__main__":
    pool = multiprocessing.Pool(processes=4)

    # Fire off multiple tasks with a callback
    # The callback handles the data automatically when ready
    for i in range(5):
        pool.apply_async(slow_square, (i,), callback=collect_result)

    print("Main process continues while workers compute...")

    # Close and join to ensure all callbacks finish before the script ends
    pool.close()
    pool.join()

    print(f"Final results in main process: {results_list}")
```

### sum(a*b): Using chunk

```python
import multiprocessing
import numpy as np

def calculate_chunk_sum_product(args):
  """Calculates the sum of element-wise products for a given chunk of a and b."""
  a_chunk, b_chunk = args
  return np.sum(a_chunk * b_chunk)

if __name__ == '__main__':
  # Generate some example data
  list_length = 100 # A large list to demonstrate parallelism
  a = np.random.rand(list_length)
  b = np.random.rand(list_length)

  num_processors = 2 # As requested by the user
  print(f"Calculating sum(a*b) using {num_processors} processors.")

  # Divide the data into chunks
  chunk_size = list_length // num_processors
  chunks_a = [a[i * chunk_size:(i + 1) * chunk_size] for i in range(num_processors)]
  chunks_b = [b[i * chunk_size:(i + 1) * chunk_size] for i in range(num_processors)]

  # Handle any remaining elements if list_length is not perfectly divisible
  if list_length % num_processors != 0:
    chunks_a[-1] = np.concatenate((chunks_a[-1], a[num_processors * chunk_size:]))
    chunks_b[-1] = np.concatenate((chunks_b[-1], b[num_processors * chunk_size:]))

  # Create a Pool of worker processes
  with multiprocessing.Poo(processes=num_processors) as pool:
    # Map the function to the chunks of data
    # pool.map distributes the (a_chunk, b_chunk) tuples to the worker function
    partial_sums = pool.map(calculate_chunk_sum_product, zip(chunks_a, chunks_b))

  # Sum the results from each processor
  total_sum = np.sum(partial_sums)

  print(f"Total sum(a*b) calculated in parallel: {total_sum}")

  # For comparison, calculate sequentially
  sequential_sum = np.sum(a * b)
  print(f"Total sum(a*b) calculated sequentially: {sequential_sum}")
  print(f"Are results close? {np.isclose(total_sum, sequential_sum)}")
```

### Shared Memory (sum(a))

```python
import numpy as np
import multiprocessing as mp
from multiprocessing import shared_memory

def worker(shm_name, dt, shp, start, end):
    # Re-attach to shared memory, create array view
    shm = shared_memory.SharedMemory(name=shm_name)
    arr = np.ndarray(shp, dtype=dt, buffer=shm.buf)

    # Sum assigned chunk
    p_sum = np.sum(arr[start:end])
    shm.close() # Detach
    return p_sum

if __name__ == "__main__":
    N = 10_000
    a = np.arange(N, dtype=np.float64)
    print(f"Seq sum: {np.sum(a)}\n")

    # Create shared memory and copy data
    shm_a = shared_memory.SharedMemory(create=True, size=a.nbytes)
    shared_a = np.ndarray(a.shape, dtype=a.dtype, buffer=shm_a.buf)
    shared_a[:] = a[:] # Copy

    # Distribute work
    n_procs = mp.cpu_count() if mp.cpu_count() > 1 else 2
    chunk = N // n_procs
    tasks = []
    for i in range(n_procs):
        s = i * chunk
        e = (i + 1) * chunk if i < n_procs - 1 else N
        tasks.append((shm_a.name, a.dtype, a.shape, s, e))

    # Parallel sum
    print(f"Starting parallel sum with {n_procs} processes...")
    with mp.Pool(processes=n_procs) as pool:
        sums = pool.starmap(worker, tasks)

    # Aggregate results
    total_sum_par = sum(sums)
    print(f"Par sum: {total_sum_par}")
    # Cleanup
    shm_a.close()
    shm_a.unlink()
```

Explanation (Simplified):

1.  **Setup**: A large NumPy array `a` is created. Its contents are copied into a `shared_memory` block (`shm_a`) accessible by all processes.
2.  **Worker Function (`worker`)**: Each worker process re-attaches to the shared memory using its `shm_name`, views it as a NumPy array (`arr`), and sums its assigned `start:end` portion.
3.  **Parallel Execution**: The main process divides the array into `n_procs` chunks and uses `mp.Pool` to run the `worker` function on each chunk in parallel.
4.  **Aggregation & Cleanup**: Partial sums from workers are combined, and the shared memory is then released.

```python
import multiprocessing as mp
import random

def monte_carlo_task(samples):
  """Heavy computation: Estimate points inside a circle."""
  hits = 0
  for _ in range(samples):
      x, y = random.random(), random.random()
      if x**2 + y**2 <= 1.0:
          hits += 1
  return hits

if __name__ == "__main__":
  total_samples = 100_000
  num_tasks = 20
  samples_per_task = total_samples // num_tasks

  print("Master process: Starting a synchronous pool...")

  with mp.Pool(processes=mp.cpu_count()) as pool:
    # Use pool.map for synchronous execution
    # It will block until all tasks are complete and return a list of results
    results = pool.map(monte_carlo_task, (samples_per_task,) * num_tasks)

  print("Master process: All tasks completed synchronously.")

  # Final reduction (HPC logic)
  pi_est = (4.0 * sum(results)) / total_samples
  print(f"Final Pi Estimate: {pi_est}")
```

```python
import numpy as np
import random
import time
import multiprocessing

def local_array_mult(args):
    """Calculates the dot product for a chunk of matrix A with the vector vect.
    Returns the partial result for the product vector.
    """
    A_chunk, vect = args # Unpack the chunk of A and the full vector
    return np.dot(A_chunk, vect)

if __name__ == "__main__":
    max_procs = 2
    m = 8000
    n = 8000

    A = np.random.random((m, n))
    vect = np.random.random(n)

    # chunk of rows for A for processor
    block_size = m // max_procs


    inputs_for_pool = []
    for i in range(max_procs):
        start_row = i * block_size
        end_row = (i + 1) * block_size
        # Ensure the last chunk takes any remaining rows if m is not perfectly divisible
        if i == max_procs - 1:
            end_row = m
        inputs_for_pool.append((A[start_row:end_row, :], vect))

    with multiprocessing.Pool(max_procs) as pool:
        partial_prods = pool.map(local_array_mult, inputs_for_pool)

    prod = np.concatenate(partial_prods)

    print("Final product (parallel calculation):")
    print(prod)

    # For verification, calculate sequentially
    sequential_prod = np.dot(A, vect)
    print("\nExpected product (sequential calculation):")
    print(sequential_prod)

    # Check if the results are numerically close
    print("\nAre parallel and sequential results close?", np.allclose(prod, sequential_prod))
```

### sum(a*b)

```python
import threading
import numpy as np
import random
import time

N = 10**5
a = np.random.random(N)
b = np.random.random(N)
max_threads = 4; sum = 0.0

threadLock = threading.Lock()

def array_mult(thread_number):
  global sum
  block_size = N//max_threads
  start_i = thread_number*block_size
  end_i = start_i + block_size
  local_sum = np.sum(a[start_i:end_i]*b[start_i:end_i])
  print(f"thread_no: {thread_number}, local_sum: {local_sum}" )
  threadLock.acquire()
  sum += local_sum
  threadLock.release()

if __name__ == "__main__":
  threads = []
  t1 = time.time()
  for i in range(max_threads):
    t = threading.Thread(target=array_mult, args=(i,))
    threads.append(t)
    t.start()
  for t in threads:
    t.join()

  print ("sum = ", sum)

  t2 = time.time()
  print ("for n threads, time = ", t2-t1)
```

```python
import threading
import random
import time # Import time module

threadLock = threading.Lock()

def monte_carlo_task(samples):
  global total_hits
  local_hits = 0 # Initialize local_hits for each thread
  for _ in range(samples):
    x, y = random.random(), random.random()
    if x**2 + y**2 <= 1.0:
      local_hits += 1
  threadLock.acquire()
  total_hits += local_hits
  threadLock.release()

if __name__ == "__main__":
  total_samples = 100_000
  max_threads = 4;
  local_samples = total_samples // max_threads
  total_hits = 0

  threads = [] # Initialize the threads list

  t1 = time.time()
  for i in range(max_threads):
    t = threading.Thread(target=monte_carlo_task, args=(local_samples,))
    threads.append(t)
    t.start()
  for t in threads:
    t.join()

  t2 = time.time()
  pi_est = (4.0 * total_hits) / total_samples
  print(f"Final Pi Estimate: {pi_est}")
  print(f"Time taken: {t2 - t1:.4f} seconds")
```

```python
import numpy as np
import threading
import time

max_threads = 4;
n = 8000; m = 8000

A = np.random.random((m,n))
vect = np.random.random(n)
prod = np.random.random(m)
block_size = m//max_threads

def array_mult(i):
  start = block_size*i
  end = start + block_size
  for i in range(start,end):
    prod[i] = np.sum(A[i,:]*vect[:])

if __name__ == "__main__":
  threads = []

  t1 = time.time()
  for i in range(max_threads):
    t = threading.Thread(target=array_mult, args=(i,))
    threads.append(t)
    t.start()
  for t in threads:
    t.join()

  t2 = time.time()
  format = " threads = %i, time diff =  %.3f"
  print(format % (max_threads, t2-t1))

```
